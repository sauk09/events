 **__init__** is a special method known as the constructor.
The __init__ method allows you to initialize the attributes (variables) of an object

The **self** parameter is a reference to the current instance of the class.
It is used to access properties and methods that belong to the class.

```python
class Bank:
  def __init__(self,name,code,email):
    self.BankName=name
    self.IFSCcode=code
    self.email=email

  def BankCode(self):
    return self.IFSCcode

Bank1=Bank('sbi',10254,'sbi@email.com')
print(Bank1.BankCode())
print(Bank.BankCode(Bank1))

# Bank1.BankAccountCode()
# Bank.BankAccountCode(Bank1)
#When we call method on the class interpreter does not know which
#instance to run the method with
```

Class variables are variables shared among all instances of class


```python
class School:
  No_of_Schools=0

  def __init__(self,name,code,email,fee):
    self.Name=name
    self.code=code
    self.email=email
    self.fee=fee

    School.No_of_Schools+=1

  def SchoolCode(self):
    return self.code+'new school'

  def fee_discount(self):
    self.fee=int(self.fee-2500)

School1=School('PrimarySchool',10254,'school@email.com',50000)
print(School1.fee)
School1.fee_discount()
print(School1.fee)
print(School1.__dict__)
print(School.No_of_Schools)
```
